```python
import tensorflow as tf
import numpy as np
from datasets import load_dataset
```

## NER Datasets

```python
def load_ner_dataset(name,subset):
  dataset = load_dataset(name,subset)
  print(dataset)
  return (
      tf.data.Dataset.from_tensor_slices((tf.ragged.constant(dataset["train"]["tokens"]),tf.ragged.constant(dataset["train"]["ner_tags"]))),
      tf.data.Dataset.from_tensor_slices((tf.ragged.constant(dataset["validation"]["tokens"]),tf.ragged.constant(dataset["validation"]["ner_tags"]))),
      tf.data.Dataset.from_tensor_slices((tf.ragged.constant(dataset["test"]["tokens"]),tf.ragged.constant(dataset["test"]["ner_tags"])))
  )
```

### Naamapadam

```python
train_ds, validation_ds, test_ds = load_ner_dataset('ai4bharat/naamapadam',"ta")
```

### Wiki Ann Dataset

```python
train_ds, validation_ds, test_ds = load_ner_dataset('wikiann',"ta")
```

### Indic-glue Wiki-ner

```python
train_ds, validation_ds, test_ds = load_ner_dataset('indic_glue',"wiki-ner.ta")
```
